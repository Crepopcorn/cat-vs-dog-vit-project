```python
import os
val_data_cat = '/content/drive/My Drive/data/val/cat'
val_data_dog = '/content/drive/My Drive/data/val/dog'
cat_items = os.listdir(val_data_cat)
dog_items = os.listdir(val_data_dog)
new_cat_items = [val_data_cat+'/' + x for x in cat_items]
new_dog_items = [val_data_dog+'/' + y for y in dog_items]
cat_num = len(cat_items)
dog_num = len(dog_items)
```

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests
import re
```

```python
a=0
b=0
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

import time
time_open = time.time()


for new_cat_item in new_cat_items:
  image = Image.open(new_cat_item)
  feature_extractor = ViTFeatureExtractor.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  model = ViTForImageClassification.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  inputs = feature_extractor(images=image, return_tensors="pt")
  tensor = str(model(**inputs).logits.argmax(dim=1))
  if (tensor != 'tensor([3])'):
    lab_val = int(re.sub('\].*','',re.sub('.*\[','',tensor)))
    print(new_cat_item + ' is ' + classes[lab_val])
    a+=1
  outputs = model(**inputs)

for new_dog_item in new_dog_items:
  image = Image.open(new_dog_item)
  feature_extractor = ViTFeatureExtractor.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  model = ViTForImageClassification.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  inputs = feature_extractor(images=image, return_tensors="pt")
  tensor = str(model(**inputs).logits.argmax(dim=1))
  if (tensor != 'tensor([5])'):
    lab_val = int(re.sub('\].*','',re.sub('.*\[','',tensor)))
    print(new_dog_item + ' is ' + classes[lab_val])
    b+=1
  outputs = model(**inputs)

time_end = time.time() - time_open
print(time_end)
```

```python
cat_val_accuracy = (cat_num - a)/cat_num*100
dog_val_accuracy = (dog_num - b)/dog_num*100
overall_val_accuracy = (cat_num + dog_num - a - b)/(cat_num + dog_num)*100
print('Validation Accuracy for cat is ' + str(cat_val_accuracy) +'%\n')
print('Validation Accuracy for dog is ' + str(dog_val_accuracy) +'%\n')
print('Overall Validation Accuracy is ' + str(overall_val_accuracy) +'%\n')
```

```python
import os
img_data = '/content/drive/My Drive/data/test'
#os.chdir(img_data)
#test_items = os.listdir(img_data)
test_items = [str(z+1)+'.jpg' for z in range(500)]
new_test_items = [img_data+'/' + x for x in test_items]
length = len(test_items)
```

```python
test_result = []
c = 1

for new_test_item in new_test_items:
  image = Image.open(new_test_item)
  feature_extractor = ViTFeatureExtractor.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  model = ViTForImageClassification.from_pretrained('nateraw/vit-base-patch16-224-cifar10')
  inputs = feature_extractor(images=image, return_tensors="pt")
  tensor = str(model(**inputs).logits.argmax(dim=1))
  if (tensor == 'tensor([3])'):
    test_result.append(0)
    print(str(c)+"\t"+str(0))
    c+=1
  elif (tensor == 'tensor([5])'):
    test_result.append(1)
    print(str(c)+"\t"+str(1))
    c+=1
  else:
    test_result.append(2)
    print(str(c)+"\t"+"Out of prediction")
    c+=1

  outputs = model(**inputs)
```

```python
import pandas as pd

name = [str(z+1) for z in range(500)]
dict = {'id': name, 'label': test_result}

df = pd.DataFrame(dict)
df.to_csv('/content/drive/My Drive/Submission.csv', index=False)
```
